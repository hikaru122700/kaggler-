データを説明変数と目的変数に分割します。今回の目的変数は「TARGET」なので、それ以外が説明変数となります。

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neural_network import MLPClassifier

import numpy as np
import gc
from itertools import combinations
from sklearn.model_selection import KFold
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from xgboost import XGBRegressor
from catboost import CatBoostRegressor

def get_models_trained(train, test, target, num_folds, train_eval):
    kf = KFold(n_splits=num_folds, shuffle=True, random_state=13)

    oof_predictions = np.zeros(len(train))
    test_predictions = np.zeros(len(test))
    train_eval_predictions = np.zeros(len(train_eval))

    # エラーの累計を保持する辞書
    sum_mse = {}
    sum_mae = {}
    # 各組み合わせの選択回数を保持する辞書
    combination_count = {}

    # 個別モデルの名前リスト
    model_names = ["model1", "model2", "model3", "model4"]

    # 全ての組み合わせを生成（1, 2, 3, 4モデル）
    all_combinations = []
    for r in range(1, 5):
        all_combinations.extend(combinations(model_names, r))

    # 初期化
    for comb in all_combinations:
        sum_mse[comb] = 0
        sum_mae[comb] = 0
        combination_count[comb] = 0

    for fold, (train_index, valid_index) in enumerate(kf.split(train, target)):
        print(f"Starting Fold {fold + 1}")
        X_train, X_valid = train[train_index], train[valid_index]
        y_train, y_valid = target[train_index], target[valid_index]

        # モデルと損失の初期化
        mse_dict = {}
        mae_dict = {}
        valid_pred_dict = {}

        # モデル1: XGBRegressor
        params1 = {
            'n_estimators': 625,
            'max_depth': 3,
            'learning_rate': 0.051151683110662566,
            'reg_alpha': 2.5636916289607576,
            'reg_lambda': 6.894373540615878,
            'subsample': 0.7499584583311859,
            'colsample_bytree': 0.7769338267922302,
            'gamma': 3.6778063980918563,
            'min_child_weight': 4
        }
        model1 = XGBRegressor(**params1)

        # モデル2: CatBoostRegressor
        params2 = {
            'iterations': 855,
            'depth': 5,
            'learning_rate': 0.05951143529700651,
            'l2_leaf_reg': 7.6281422102055005,
            'border_count': 180,
            'subsample': 0.664450188123631,
            'colsample_bylevel': 0.8000552761878679,
            'random_strength': 0.26553215110853756,
            'bagging_temperature': 0.6197718091674634,
            'random_seed': 13,
            'verbose': 0,
        }
        model2 = CatBoostRegressor(**params2)

        # モデル3: XGBRegressor
        params3 = {
            'n_estimators': 775,
            'max_depth': 4,
            'learning_rate': 0.059967058989802736,
            'reg_alpha': 4.252608756868103,
            'reg_lambda': 6.872086297579257,
            'subsample': 0.7146820344108187,
            'colsample_bytree': 0.7295963801421386,
            'gamma': 3.9044476288195145,
            'min_child_weight': 10
        }
        model3 = XGBRegressor(**params3)

        # モデル4: CatBoostRegressor
        params4 = {
            'iterations': 958,
            'depth': 4,
            'learning_rate': 0.06281504158226449,
            'l2_leaf_reg': 6.412129799918053,
            'border_count': 216,
            'subsample': 0.5878822652416278,
            'colsample_bylevel': 0.8341498722019399,
            'random_strength': 2.2737727901801135,
            'bagging_temperature': 0.9376789810514898,
            'random_seed': 13,
            'verbose': 0,
        }
        model4 = CatBoostRegressor(**params4)

        # 各モデルの訓練と予測
        models = {
            "model1": model1,
            "model2": model2,
            "model3": model3,
            "model4": model4
        }

        for name, model in models.items():
            model.fit(X_train, y_train)
            preds = model.predict(X_valid)
            mse = mean_squared_error(y_valid, preds)
            mae = mean_absolute_error(y_valid, preds)
            print(f"Fold {fold + 1} MSE for {name} = {mse}")
            print(f"Fold {fold + 1} MAE for {name} = {mae}")
            mse_dict[(name,)] = mse  # MSEを保存
            mae_dict[(name,)] = mae    # MAEを保存
            valid_pred_dict[(name,)] = preds
            sum_mse[(name,)] += mse
            sum_mae[(name,)] += mae

        # すべての2モデルおよび3モデルの組み合わせを評価
        for r in [2, 3]:
            for comb in combinations(model_names, r):
                # 平均予測
                preds_comb = np.mean([valid_pred_dict[(comb_part,)] for comb_part in comb], axis=0)
                mse_comb = mean_squared_error(y_valid, preds_comb)
                mae_comb = mean_absolute_error(y_valid, preds_comb)
                print(f"Fold {fold + 1} MSE for {comb} = {mse_comb}")
                print(f"Fold {fold + 1} MAE for {comb} = {mae_comb}")
                mse_dict[comb] = mse_comb
                mae_dict[comb] = mae_comb
                valid_pred_dict[comb] = preds_comb
                sum_mse[comb] += mse_comb
                sum_mae[comb] += mae_comb

        # 全モデルのアンサンブル
        preds_all = np.mean([valid_pred_dict[(model,)] for model in model_names], axis=0)
        mse_all = mean_squared_error(y_valid, preds_all)
        mae_all = mean_absolute_error(y_valid, preds_all)
        print(f"Fold {fold + 1} MSE for all models ensemble = {mse_all}")
        print(f"Fold {fold + 1} MAE for all models ensemble = {mae_all}")
        mse_dict[tuple(model_names)] = mse_all
        mae_dict[tuple(model_names)] = mae_all
        valid_pred_dict[tuple(model_names)] = preds_all
        sum_mse[tuple(model_names)] += mse_all
        sum_mae[tuple(model_names)] += mae_all

        # 最適な組み合わせを選択（MSEが最小）
        # もしくは他の基準に基づいて選択
        # ここではMSEが最小の組み合わせを選択
        best_combination = min(mse_dict, key=mse_dict.get)  # MSEが最小の組み合わせを選択
        best_mse = mse_dict[best_combination]
        best_mae = mae_dict[best_combination]
        print(f"Fold {fold + 1} best combination: {best_combination} with MSE = {best_mse} and MAE = {best_mae}")

        # OOF予測とテスト予測の更新
        if isinstance(best_combination, tuple):
            # 複数モデルのアンサンブル
            oof_predictions[valid_index] = valid_pred_dict[best_combination]
            test_preds = np.mean([models[model].predict(test) for model in best_combination], axis=0)
            train_eval_preds = np.mean([models[model].predict(train_eval) for model in best_combination], axis=0)
        else:
            # 単一モデル（この場合は常にタプルになるため不要）
            oof_predictions[valid_index] = valid_pred_dict[(best_combination,)]
            test_preds = models[best_combination].predict(test)
            train_eval_preds = models[best_combination].predict(train_eval)

        # 各組み合わせの選択回数をインクリメント
        combination_count[best_combination] += 1

        test_predictions += test_preds / kf.n_splits
        train_eval_predictions += train_eval_preds / kf.n_splits

        # メモリの解放
        del X_train, X_valid, y_train, y_valid, model1, model2, model3, model4
        gc.collect()

        print('---------------\n')

    final_mse = mean_squared_error(target, oof_predictions)
    final_mae = mean_absolute_error(target, oof_predictions)
    print(f"OOF MSE = {final_mse}")
    print(f"OOF MAE = {final_mae}")

    # 各組み合わせの累積MSEとMAEを表示
    print("Sum of MSE for each combination:")
    for comb, mse in sum_mse.items():
        print(f"{comb}: {mse}")

    print("\nSum of MAE for each combination:")
    for comb, mae in sum_mae.items():
        print(f"{comb}: {mae}")

    print("\nNumber of times each combination was selected:")
    for comb, count in combination_count.items():
        print(f"{comb}: {count}")

    return oof_predictions, test_predictions, train_eval_predictions
```

```python
# 目的変数と説明変数に分割
X = train_all.iloc[:, :-1].values
y = train_all.iloc[:, -1].values
X_test = test.values
```

```python
oof_predictions, test_preds, train_preds = get_models_trained(X, X_test, y, 5, X)
```
